### Formatting date  to unifde Format "%m-%d-%y"

```python
# Formatting date  to unifde Format "%m-%d-%y"the date in data represented in tow format '6/29/1984'  with hyphen
# (/) and '6/29/1984'  with forward slash (-)
```

```python
import re
from datetime import datetime


regex1 = r'\d{1,2}/\d{1,2}/\d{4}' # date like '6/29/1984'  with hyphen (/)

# regex2 = r'\d{1,2}-\d{1,2}-\d{2}' # date like '6/29/1984'  with forward slash (-)


for index in range(len(HR_Data['birth_date'])):
    
    date_ = HR_Data['birth_date'][index]
    
    if re.match(regex1, date_):
        
#         print(HR_Data['birth_date'][index])
        
        date_obj = datetime.strptime(date_, "%m/%d/%Y")

        formatted_date = date_obj.strftime("%Y-%-m-%-d")


        
        HR_Data['birth_date'][index] = formatted_date

    
    else : 

        date_obj = datetime.strptime(date_, "%m-%d-%y")


        formatted_date = date_obj.strftime("%Y-%-m-%-d")

        
        HR_Data['birth_date'][index] = formatted_date    
    
```

### Addig the age column to data 

```python
from datetime import date
 
def calculateAge(birth_Date):
    
    today = date.today()
    
    age = today.year - birth_Date.year - ((today.month, today.day) < (birth_Date.month, birth_Date.day))
 
    return age

print(calculateAge(date(1997, 2, 3)), "years")
```

```python
# Apply the calculateAge function to the birth_date column

HR_Data['age'] = HR_Data['birth_date'].apply(calculateAge)
```
